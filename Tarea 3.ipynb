```python
#1. Calculadora Simple

suma = "suma" #Definición de variables
resta = "resta"
multiplicacion = "multiplicacion"
division = "division"

operacion = input("Escriba qué operación quiere realizar en minúsculas sin acentos: ")

if operacion==suma: #Casos de la calculadora
  num_uno = float(input("Escriba el primer sumando: "))
  num_dos = float(input("Escriba el segundo sumando: "))
  resultado = num_uno + num_dos
  print(num_uno, " mas ", num_dos, " es ", resultado)
elif operacion==resta:
  num_uno = float(input("Escriba el minuendo: "))
  num_dos = float(input("Escriba el sustraendo: "))
  resultado = num_uno - num_dos
  print(num_uno, " menos ", num_dos, " es ", resultado)
elif operacion==multiplicacion:
  num_uno = float(input("Escriba el multiplicando: "))
  num_dos = float(input("Escriba el multiplicador: "))
  resultado = num_uno * num_dos
  print(num_uno, " por ", num_dos, " es ", resultado)
elif operacion==division:
  num_uno = float(input("Escriba el dividendo: "))
  num_dos = float(input("Escriba el divisor: "))
  resultado = num_uno / num_dos
  print(num_uno, " entre ", num_dos, " es ", resultado)
else:
  print("Por favor, escriba una operación.")
```

```python
#6. Cálculo del Promedio}

lista = input("Escriba los números a promediar separados por comas: ")
numeros = lista.split(", ") #Excluir las comas y espacios de la lista
nueva_lista = [float(num) for num in numeros] #Redefinición del tipo de dato de la lista
promedio = sum(nueva_lista) / len(nueva_lista)
print("El promedio de los números en su lista es ", promedio)
```

```python
#10. Registro de Estudiantes

estudiantes = input("Escriba los nombres de sus estudiantes separados por comas: ")
nombres = estudiantes.split(", ")
for nombre in nombres:
  print("Escriba las calificaciones de ", nombre, "separadas por comas: ")
  calificaciones = input()
  lista_cal = calificaciones.split(", ")
  listaCal = [float(num) for num in lista_cal]
  promedio = sum(listaCal) / len(listaCal)
  print("El promedio de ", nombre, "es", promedio)
```

```python
#3. Comentarios Informativos

#Programa para encontrar el número máximo en una lista

def encontrar_maximo(numeros):
    if not numeros: # Verificamos si la lista está vacía
        return "La lista está vacía."
    maximo = numeros[0]  # Inicializamos 'maximo' con el primer elemento de la lista
    for num in numeros:# Iteramos sobre cada número en la lista
        if num > maximo:# Comparamos el número actual con el máximo encontrado hasta ahora
            maximo = num  # Actualizamos 'maximo' si encontramos un número mayor
    return maximo  # Devolvemos el número máximo encontrado
# Ejemplo de uso
lista_numeros = [-3, -5, -2, -8, -1, -10, -4]
resultado = encontrar_maximo(lista_numeros)
print(f"El número máximo en la lista es: {resultado}")
```

```python
#6. Creación de un Programa con Comentarios de Uso.

# Programa para crear una lista de supermercado
# Este programa permite a los usuarios agregar artículos a una lista y ver la lista.
# Para usarlo, simplemente sigue las instrucciones que aparecen en la consola.

def main():
    lista_supermercado = []  # Inicializamos una lista vacía
    while True:
        articulo = input("Ingresa un artículo para agregar a la lista (o 'salir' para terminar): ")
        if articulo.lower() == 'salir':
            break  # Salimos del bucle si el usuario escribe 'salir'
        lista_supermercado.append(articulo)  # Agregamos el artículo a la lista
        print(f"{articulo} ha sido agregado a la lista.")
    # Mostrar la lista final de artículos
    print("\nTu lista de supermercado:")
    for item in lista_supermercado:
        print(f"- {item}")
# Ejecutar el programa
if __name__ == "__main__":
    main()
```
